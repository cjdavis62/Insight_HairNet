```python
import instaloader
from instaloader import Instaloader, Profile
from itertools import islice
```

```python
USER="cjdavis_physics"
```

```python
L = instaloader.Instaloader(sleep=True, quiet=False, user_agent=None, dirname_pattern=None, filename_pattern=None,
                            download_pictures=True, download_videos=False, download_video_thumbnails=False, download_geotags=False,
                            download_comments=True, save_metadata=False, compress_json=True, post_metadata_txt_pattern=None,
                            storyitem_metadata_txt_pattern=None, max_connection_attempts=3, commit_mode=False)
L.interactive_login(USER)
```

```python
list_of_instagram_names=["urbanbettysalon", "methodhair", "redstellasalon", "topazsalonaustin", "garboasalon", "frenchysbeauty",
                        "blackorchidsalon", "cnnhairteam", "acessalon", "benjaminbeausalon", "vainaustin", "loveandrootssalon", "wildorchidatx",
                        "salonsovayatx", "thesalonatthedomain", "ritualsalonatx", "bellasalonatx", "milkandhoneysalon", "pathsalon", "waterstone_salon"]
print(len(list_of_instagram_names))
```

```python
%%timeit
number_downloaded = 0
for name in list_of_instagram_names:
    print(name)
    profile = Profile.from_username(L.context, name)
    print(profile.external_url)
    for post in islice(profile.get_posts(), 50):
        number_downloaded = number_downloaded + 1
        L.download_post(post, profile)
        print (str(number_downloaded) + " photo(s) downloaded")
```
