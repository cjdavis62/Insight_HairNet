# Now deal with the salon data and create a csv with the relevant information

```python
salon_path = "/home/cjdavis/insight/instaloader/salons/"
salon_insta_names = os.listdir(salon_path)
salon_photos_path_list = []
for salon_name in salon_insta_names:
    salon_photos_path = salon_path + salon_name + "/"
    salon_photos_path_list.append(salon_photos_path)
    print(salon_photos_path)
```

```python
photo_location_list = []
photo_list = []
for salon_path in salon_photos_path_list:
    photos = os.listdir(salon_path)
    for photo in photos:
        if ".jpg" in photo:
            photo_location_list.append(salon_path)
            photo_list.append(photo)
```

```python
insta_df = pd.DataFrame(list(zip(photo_list, photo_location_list)), columns = ["photo_path", "salon_local_path"])
```

```python
def get_instagram_name(row):
    instagram_name = row["salon_local_path"].split("/")[6]
    return (instagram_name)
def get_salon_name_from_insta_name(insta_name, dict_of_names):
    salon_name = dict_of_names[insta_name]
    return (salon_name)
```

```python
insta_df["instagram_name"] = insta_df.apply(get_instagram_name, axis=1)
```

```python
insta_df['salon_name'] = insta_df['instagram_name'].apply(lambda x: get_salon_name_from_insta_name(x, dict_of_names))
```

## load data into database ##

```python
f=open("db.txt", "r")
contents=0
if not(contents):
    contents = f.read()
username = contents.split("\n")[0]
password = contents.split("\n")[1]
dbname = 'salons_db'
```

```python
engine = create_engine('postgres://%s:%s@localhost/%s'%(username,password, dbname))
print(engine)
```

```python
if not database_exists(engine.url):
    create_database(engine.url)
print(database_exists(engine.url))
```

```python
con = psycopg2.connect(database = dbname, user = username, password = password, port=5432, host= "/var/run/postgresql/")
```

```python
insta_df.to_sql('insta_data_table', engine, if_exists='replace')
con.close()
```
